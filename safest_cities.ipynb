```python
# Load the data
df = pd.read_csv('Resources/df_cleaned.csv')

# Initialize an empty map centered on the United States (or an average location)
map = folium.Map(location=[37.0902, -95.7129], zoom_start=5)  # Central USA coordinates

# Function to add markers for each year
def add_markers_by_year(year):
    # Filter the data for the specific year
    yearly_data = df[df['report_year'] == year]
    
    # Find the safest and most dangerous city based on crimes_percapita
    safest_city = yearly_data.loc[yearly_data['crimes_percapita'].idxmin()]
    most_dangerous_city = yearly_data.loc[yearly_data['crimes_percapita'].idxmax()]
    
    # Safest city marker
    folium.Marker(
        location=[safest_city['latitude'], safest_city['longitude']],
        popup=f"{safest_city['agency_jurisdiction']} - Safest City",
        tooltip="Safest City",
        icon=folium.Icon(color="blue")
    ).add_to(map)

    # Most dangerous city marker
    folium.Marker(
        location=[most_dangerous_city['latitude'], most_dangerous_city['longitude']],
        popup=f"{most_dangerous_city['agency_jurisdiction']} - Most Dangerous City",
        tooltip="Most Dangerous City",
        icon=folium.Icon(color="red")
    ).add_to(map)

# Loop through each unique year and add markers
for year in df['report_year'].unique():
    add_markers_by_year(year)

# Save or display the map
map.save("crime_map.html")
map
```
